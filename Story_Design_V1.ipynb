```
import pandas as pd
import numpy as np
from openpyxl.workbook import Workbook
from openpyxl import load_workbook
```

```
df_example = pd.read_excel('Test_Example.xlsx', sheet_name = 'Example', dtype="string")
df_example.dropna(subset=['Scenario'], inplace = True)
df_example.drop([0], inplace = True)
df_example = df_example.astype('string')
df_example
```

```
events = df_example['Event'].dropna()
events
```

```
df_example['Result'] = df_example[df_example.columns[3:]].apply(
    lambda x: '   '.join(x.dropna()),
    axis=1
)
df_example
```

```
df_example['Event'].fillna(method='ffill', inplace = True)
df_example
df_out = df_example[['Event','Result']]
df_out
```

```
temp = df_out.iloc[0,0]
entry = df_out.iloc[0,1]
l = len(df_example['Event'])
entries = []

for i in range(1, l):
  if df_out.iloc[i, 0] == temp:
    entry = '\n'.join([entry,df_out.iloc[i,1]])
  else:
    temp = df_out.iloc[i, 0]
    entries.append(entry)
    entry = df_out.iloc[i,1]
entries.append(entry)
ds = pd.Series(entries)
ds
```

```
df_design = pd.read_excel('Test_Example.xlsx', sheet_name = 'Design', dtype="string")
df_final = pd.DataFrame()

for counter in range(len(events)):
  df_temp = df_design.copy()
  df_temp.iloc[0,1] += str(counter+1)
  df_temp.iloc[0,2] += events.iloc[counter]
  df_temp.iloc[0,1] += df_temp.iloc[0,2]
  df_temp.iloc[0,3] += events.iloc[counter]
  df_temp.iloc[5,7] = df_temp.iloc[5,7] + "\n" + ds.iloc[counter]
  frames = [df_final, df_temp]
  df_final =  pd.concat(frames)
  
df_final.drop(columns=["Test Case Name R"], inplace = True)
df_final['Test Case Name'].fillna(method='ffill', inplace = True)
df_final['Description'].fillna(method='ffill', inplace = True)

df_final.to_excel("output.xlsx",sheet_name='Ex314', index=False) 
```
